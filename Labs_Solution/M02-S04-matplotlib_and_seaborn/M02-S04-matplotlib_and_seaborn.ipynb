### From the fitbit dataframe, we want to visually understand:
- Average number of steps by month. Use the appropriate visualization to show the `median` `steps` by `month`.
- Is Fitbitter more active on weekend or workdays? 
- All plots must be in the same jupyter notebook cell.
- Try to create titles and label your axis. 

**Hints**:
- Use `Steps`, `Months_encoded` and `Work_or_Weekend` columns.
- Set your size figure to 12,4
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures. 

```python
# Your code here
```

```python
#Average number of steps by month
month_steps = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()
month_steps
```

```python
# Use during weekend or workdays
weekday_steps = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median().sort_values()
weekday_steps
```

```python
fig,axes = plt.subplots(figsize = (12,4), nrows = 1, ncols = 2)

#Average number of steps by month
plt.sca(axes[0])
month_steps.plot(kind = 'line',linewidth=2)

plt.ylabel('Median number of steps')
plt.xlabel('Months')
plt.title('Median Steps walked by month')
plt.xticks(list(range(1,13)),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

plt.savefig('steps_months.png',bbox_inches='tight')


# Use during weekend or workdays
plt.sca(axes[1])
weekday_steps.plot(kind = 'bar',alpha = 0.5)
plt.ylabel('Median number of steps')
plt.title('Median Steps walked by Workdays and Weekend')
plt.xticks(list(range(2)),['Weekend','Workdays'])

plt.savefig('steps_work_weekend.png',bbox_inches='tight')

plt.show()
```

### Plot 3 scatter plots of the following features:
- Minutes Lightly Active vs Steps
- Minutes Very Active vs Steps
- Minutes Sedentary vs Steps

```python
# Your code here
```

```python
# Your code here#Method 1
fig,axes = plt.subplots(figsize = (16,8), nrows = 1, ncols = 3)
plt.sca(axes[0])
plt.scatter(fitbit['Minutes Lightly Active'],fitbit['Steps'], alpha = 0.8)
plt.xlabel('Minutes Lightly Active')
plt.ylabel('Steps')
    
plt.sca(axes[1])
plt.scatter(fitbit['Minutes Very Active'],fitbit['Steps'], alpha = 0.8)
plt.xlabel('Minutes Very Active')

plt.sca(axes[2])
plt.scatter(fitbit['Minutes Sedentary'],fitbit['Steps'], alpha = 0.8)
plt.xlabel('Minutes Sedentary')
```

```python
#Method 2
cols=['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']
df=fitbit[cols]
fig,axes = plt.subplots(figsize = (16,8), nrows = 1, ncols = 3)
for i in range(3):
    plt.sca(axes[i])
    plt.scatter(df.iloc[:,i],df['Steps'], alpha = 0.8)
    plt.xlabel(df.iloc[:,i].name)
    plt.ylabel('Steps')
```
